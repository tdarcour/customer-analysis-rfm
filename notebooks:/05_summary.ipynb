```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customer_agg_df = pd.read_csv('../data/customer_aggregates.csv')
rfm_df = pd.read_csv('../data/rfm_clients.csv')

customer_agg_df.head()
```

## Résultats clés

Les analyses montrent une forte concentration du chiffre d'affaires sur une minorité de clients.
Une majorité de clients dépense peu, tandis qu'un petit groupe contribue très fortement aux revenus.

```python
customer_agg_df['TotalSpent'].hist(bins=100)
plt.title('Distribution des dépenses clients')
plt.xlabel("Montant total dépensé")
plt.ylabel("Nombre de clients")
plt.show()
```

Ce graphique illustre un phénomène de type 'Long Tail' : la plupart des clients contribuent de manière marginale, tandis qu'une minorité de clients à forte valeur génère une part importante du chiffre d'affaires. Cela justifie une segmentation client afin d'adapter les actions marketing.


## Segmentation clients (RFM)

La segmentation repose sur :

- Recency : Nombre de jours depuis la dernière commande
- Frequency : Nombre de commandes
- Monetary : Montant total dépensé

Elle permet d'identifier :

- Des clients à forte valeur (Monetary élevé)
- Des clients fidèles (Frequency élevé)
- Des clients à risque (Recency élevé)

## Recommandations business

- Construire des actions marketing différenciées selon les segments identifiés.
